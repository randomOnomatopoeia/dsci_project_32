Data Description: 

players.csv 
- experience (string): Level of experience -- Beginner, Amateur, Regular, Pro, Veteran
- subscribe (boolean): whether or not they are subscribed
- hashedEmail (string): their hashed email
- played_hours (float): the number of hours they play in a week (presumably)
- name (string): Their first name
- gender (string): Their gender -- Male, Female, Non-binary, Prefer not to say
- age (int): Their age in years
- individualId (None): Empty column
- organizationName (None): Empty column


sessions.csv
- hashed email (string): their email, but hashed
- start_time (Object): the starting time with regards to a date
- end_time (object): the ending time in a date format
- original_start_time (float): a numerical representation of the date
- original_end_time (float): a numerical representation of a date
